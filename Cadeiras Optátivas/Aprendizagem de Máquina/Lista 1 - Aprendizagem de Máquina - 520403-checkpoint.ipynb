# a) Código para o OLS (mínimos quadrados ordinários)

```python
X = dados[:, 0]
y = dados[:, 1]

def OLS(x, y):
    med = np.mean(x)
    x_aux = np.std(x, ddof=1)
    X_normal = (x - med) / x_aux
    
    X_Matriz = np.c_[np.ones((X_normal.shape[0], 1)), X_normal]

    w = np.linalg.solve((X_Matriz.T @ X_Matriz), X_Matriz.T @ y)

    y_pred = X_Matriz @ w

    MSE = ((((y - y_pred) ** 2).sum()) / (X_normal.shape[0]))

    return w, MSE, y_pred

w , MSE , y_pred = OLS (X,y)

print('Parâmetros do modelo (w):', w)
print('MSE:', MSE)
```

# b) Código para o GD(Gradiente Descendente):

```python
X = dados[:, 0]
y = dados[:, 1]

def GD(x, y, alpha=0.1, max_iter=500):
    
    med = np.mean(x)
    x_aux = np.std(x, ddof=1)
    x_normal = (x - med) / x_aux

    x_matriz = np.c_[np.ones((x_normal.shape[0], 1)), x_normal.reshape(-1, 1)]

    w = np.zeros((2, 1))

    mse_values = []

    for _ in range(max_iter):

        y_pred = x_matriz @ w
        e = y - y_pred.flatten()

        w[0] = w[0] + alpha * (1/x.shape[0]) * np.sum(e)
        w[1] = w[1] + alpha * (1/x.shape[0]) * np.sum(e * x)

        mse = np.mean(e**2)
        mse_values.append(mse)

    return w, mse_values[-1] , y_pred , mse_values

w, MSE, y_pred, mse_values = GD (X,y)
print("Parâmetros do modelo (w):")
print(w)
print("Erro médio quadrático (MSE):", MSE)
```

# Código para o SGD(Gradiente Descendente Estocástico) :

```python
X = dados[:, 0]
y = dados[:, 1]

def SGB(x, y, alpha=0.1, max_iter=600):
    x_matriz = np.c_[np.ones((x.shape[0], 1)), x.reshape(-1, 1)]
    
    w = np.random.normal(0, 1, x_matriz.shape[1])
    
    idx = np.arange(len(y))
    
    mse_values = []

    for _ in range(max_iter):
        np.random.shuffle(idx)
        
        for i in idx:
            y_pred = x_matriz[i] @ w
            
            e = y[i] - y_pred

            for j in range(x_matriz.shape[1]):
                w[j] = w[j] + alpha * e * x_matriz[i,j]

        y_pred_all = x_matriz @ w

        mse = np.mean((y - y_pred_all.flatten())**2)
        mse_values.append(mse)

    return w, mse_values[-1], y_pred_all, mse_values

w, MSE, y_pred_all, mse_values = SGB(X, y)

print("Parâmetros do modelo (w):")
print(w)
print("Erro médio quadrático (MSE):", MSE)
```

# 2° Questão:

```python
caminho_arquivo2 = 'C:\\Users\\user\\Desktop\\california.csv'

dados2 = np.loadtxt(caminho_arquivo2, delimiter=',')

# Aqui vou criar um dataframe com os dados de X e y
df_2 = pd.DataFrame(dados2, columns=['Atributo1', 'Atributo2', 'Atributo3', 'Atributo4', 'Atributo5', 'Atributo6', 'Atributo7', 'Atributo8', 'Saida'])

# exibindo a tabela
display(df_2)

X = dados2[:,:-1]
y = dados2[:,-1].reshape(-1,1)
```

## Normalizando

```python
def Min_Max(X, y):
    data = np.c_[X, y]
    mini = np.min(data)
    maxi = np.max(data)
    X_norm = (X - mini)/(maxi - mini)
    y_norm = (y - mini)/(maxi - mini)
    return X_norm, y_norm, mini, maxi

def desnormalize_MSE(MSE_norm, mini, maxi):
    RMSE_desnormalizado = (MSE_norm * (maxi - mini)) + mini
    return RMSE_desnormalizado
```

## Preparando os dados

```python
def train_test_split(x, y, test_percent=0.2):
    n = int(len(x)*test_percent)
    shuffle = range(len(x))
    test_idx = random.sample(shuffle,n)
    train_idx = [i for i in shuffle if i not in test_idx]
    x_train = np.array([x[i] for i in train_idx])
    x_test = np.array([x[i] for i in test_idx])
    y_train = np.array([y[i] for i in train_idx])
    y_test = np.array([y[i] for i in test_idx])
    return x_train, x_test, y_train, y_test

def non_linear_transf(x, degree):
    if degree < 2:
        return x
    X_aux = x.copy()
    for i in range(2, degree + 1):
        for j in range(x.shape[1]):
            add_column = x[:, j] ** i
            X_aux = np.c_[X_aux, add_column]
    return X_aux

def OLS2(x, y, l2): # Pois agora vamos usar Regularização L2
    med = np.mean(x)
    x_aux = np.std(x, ddof=1)
    X_normal = (x - med) / x_aux

    I = np.eye(X_normal.shape[1] + 1)

    X_Matriz = np.c_[np.ones((X_normal.shape[0], 1)), X_normal]

    w = np.linalg.solve((X_Matriz.T @ X_Matriz) + l2 * I, X_Matriz.T @ y)

    y_pred = X_Matriz @ w

    MSE = ((((y - y_pred) ** 2).sum()) / (X_normal.shape[0]))

    return w, MSE, y_pred
```

## Treinando o modelo

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_percent=0.2)

train_errors = np.ones(13)
test_errors = np.ones(13)

for degree in range(1, 14):
    X_train_polinomial = non_linear_transf(X_train, degree)
    X_train_polinomial_norm, y_train_norm, min_train, max_train = Min_Max(X_train_polinomial, y_train)
    X_test_polinomial = non_linear_transf(X_test, degree)
    X_test_polinomial_norm, y_test_norm, min_test, max_test = Min_Max(X_test_polinomial, y_test)
    
    w, mse,y_pred = OLS2(X_train_polinomial_norm, y_train_norm, l2=0.01)
    X_test_polinomial = np.c_[np.ones(len(y_test)), X_test_polinomial]
    
    RMSE1 = np.sqrt(abs(desnormalize_MSE(mse, min_train, max_train)))
    train_errors[degree-1] = RMSE1
    test_errors[degree-1] = np.sqrt(abs((y_test - (X_test_polinomial @ w))*2).sum()/(2*(X_test.shape[0])))
```

## Resultados

```python
RMSES = [train_errors, test_errors]

print("RMSE de treino:")
print(train_errors)

print("RMSE de teste:")
print(test_errors)
```

## Visualização

```python
x_vis = [i + 1 for i in range(len(train_errors))]

plt.plot(x_vis, train_errors, marker='o')
plt.plot(x_vis, test_errors, marker='p', color='red')

plt.xlabel('Grau do Polinômio')
plt.ylabel('RMSE')
plt.title('RMSE vs Grau do Polinômio')

fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(12, 6))
ax1.set_title('RMSE x Grau do polinômio (Treino)')
ax2.set_title('RMSE x Grau do polinômio (Teste) (Escala log.)')
ax2.set_yscale('log')
ax1.plot(range(len(RMSES[0])), RMSES[0])
ax2.plot(range(len(RMSES[1])), RMSES[1])
plt.show()
```
