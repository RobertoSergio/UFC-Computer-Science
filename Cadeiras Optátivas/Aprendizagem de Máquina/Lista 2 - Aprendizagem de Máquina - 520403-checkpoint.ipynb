## Funções que serão usadas nos métodos:

```python
# ler o CSV
def csv(file):
    data = pd.read_csv(file)
    X = data.iloc[:, :-1].values
    y = data.iloc[:, -1].values
    return X, y

# FUNÇÃO PARA normalizar os dados
def normalizar_dados(X):
    return (X - np.mean(X, axis=0)) / np.std(X, axis=0)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Função de validação cruzada em 10 folds
def kfold_10(X, y):
    k = 10
    indices = np.arange(X.shape[0])
    np.random.shuffle(indices)
    tamanho_fold = len(indices) // k
    return [indices[i * tamanho_fold:(i + 1) * tamanho_fold] for i in range(k)]
```

# Código para o modelo de Regressão logística com GD

## a)  Considerando uma validação cruzada em 10 folds, avalie modelos de classificação binária nos dados em questão. Para tanto, use as abordagens abaixo:

```python
# Função de Gradiente Descendente
def GD(X, y, alpha, num_itera):
    w = np.zeros(X.shape[1])
    for _ in range(num_itera):
        n = len(y)
        h = sigmoid(X.dot(w))
        w -= alpha * (1 / n) * X.T.dot(h - y)
    return w

# Função principal da Regressão Logística
def regressao_logistica(file, alpha=0.01, num_itera=1000):

    X, y = csv(file)
    X = normalizar_dados(X)


    X = np.c_[np.ones((X.shape[0], 1)), X]

    Indice_KF = kfold_10(X, y)
    acuracias_Regressão = []
    acuracias_classe_0 = []
    acuracias_classe_1 = []

    for i in range(10):
        test_index = Indice_KF[i]
        train_index = np.concatenate([Indice_KF[j] for j in range(10) if j != i])
        
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        # Treinamento do modelo
        w = GD(X_train, y_train, alpha, num_itera)
        
        # Avaliação do modelo
        y_pred = sigmoid(X_test.dot(w))
        y_pred = np.round(y_pred).astype(int)
        aux_acuracia = np.mean(y_pred == y_test)
        acuracias_Regressão.append(aux_acuracia)
                           
        # Acurácia por classe
        acuracia_classe_0 = np.mean(y_pred[y_test == 0] == 0)
        acuracia_classe_1 = np.mean(y_pred[y_test == 1] == 1)
        acuracias_classe_0.append(acuracia_classe_0)
        acuracias_classe_1.append(acuracia_classe_1)
    
    # Calcular e exibir os resultados       
    media_acuracia_logistica = np.mean(acuracias_Regressão)
    desvio_acuracia_logistica = np.std(acuracias_Regressão)
    media_acuracia_logistica_classe_0 = np.mean(acuracias_classe_0)
    desvio_acuracia_logistica_classe_0 = np.std(acuracias_classe_0)
    media_acuracia_logistica_classe_1 = np.mean(acuracias_classe_1)
    desvio_acuracia_logistica_classe_1 = np.std(acuracias_classe_1)
    
    print('Regressão Logística:')
    print(f'{"-"*65}')
    print(f'Acurácia média: {media_acuracia_logistica}')
    print(f'Desvio padrão da acurácia: {desvio_acuracia_logistica}')
    print(f'{"-"*65}')
    print(f'Acurácia média para a classe 0: {media_acuracia_logistica_classe_0}')
    print(f'Desvio padrão da acurácia para a classe 0: {desvio_acuracia_logistica_classe_0}')
    print(f'{"-"*65}')
    print(f'Acurácia média para a classe 1: {media_acuracia_logistica_classe_1}')
    print(f'Desvio padrão da acurácia para a classe 1: {desvio_acuracia_logistica_classe_1}')
    print(f'{"-"*65}')
    
    # Plotar gráfico
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, 11), acuracias_Regressão, marker='o', linestyle='-', color='b')
    plt.xlabel('Fold')
    plt.ylabel('Acurácia')
    plt.title('Acurácia Média em cada Fold')
    plt.grid(True)
    plt.show()
```

# Código para o modelo da Análise do discriminante Gaussiano

## a)  Considerando uma validação cruzada em 10 folds, avalie modelos de classificação binária nos dados em questão. Para tanto, use as abordagens abaixo:

```python
# Função para calcular parâmetros gaussianos
def calcular_parametros_gaussianos(X, y):
    phi = np.mean(y)
    mu_0 = np.mean(X[y == 0], axis=0)
    mu_1 = np.mean(X[y == 1], axis=0)
    sigma_0 = np.cov(X[y == 0], rowvar=False)
    sigma_1 = np.cov(X[y == 1], rowvar=False)
    sigma = (y == 0).mean() * sigma_0 + (y == 1).mean() * sigma_1
    return phi, mu_0, mu_1, sigma

# Função de discriminação gaussiana
def discriminante_gaussiano(X, phi, mu_0, mu_1, sigma):
    sigma_inv = np.linalg.inv(sigma)
    p_0 = np.log(1 - phi) - 0.5 * np.sum((X - mu_0).dot(sigma_inv) * (X - mu_0), axis=1)
    p_1 = np.log(phi) - 0.5 * np.sum((X - mu_1).dot(sigma_inv) * (X - mu_1), axis=1)
    
    res = []
    for i in range(len(p_1)):
        if p_1[i] > p_0[i]:
            res.append(1)
        else:
            res.append(0)
    
    return np.array(res)

# Função principal da Análise Discriminante Gaussiana
def analise_discriminante_gaussiano(file):

    X, y = csv(file)
    X = normalizar_dados(X)
    
    Indice_KF = kfold_10(X, y)
    acuracias_Análise = []
    acuracias_classe_1 = []
    acuracias_classe_0 = []

    for i in range(10):
        test_index = Indice_KF[i]
        train_index = np.concatenate([Indice_KF[j] for j in range(10) if j != i])
        
        X_treino, X_teste = X[train_index], X[test_index]
        y_treino, y_teste = y[train_index], y[test_index]
        
        # Calcular parametros
        phi, mu_0, mu_1, sigma = calcular_parametros_gaussianos(X_treino, y_treino)
        
        # Prever com o discriminante gaussiano
        y_pred = discriminante_gaussiano(X_teste, phi, mu_0, mu_1, sigma)
        acuracia = np.mean(y_pred == y_teste)
        acuracias_Análise.append(acuracia)
        
        # Acurácia por classe
        aux_acuracias_classe_1 = np.mean(y_pred[y_teste == 1] == 1)
        aux_acuracias_classe_0 = np.mean(y_pred[y_teste == 0] == 0)
        acuracias_classe_1.append(aux_acuracias_classe_1)
        acuracias_classe_0.append(aux_acuracias_classe_0)

    # Exibir resultados
    media_acuracia_analise = np.mean(acuracias_Análise)
    desvio_acuracia_analise = np.std(acuracias_Análise)
    media_acuracia_analise_classe_0 = np.mean(acuracias_classe_0)
    desvio_acuracia_analise_classe_0 = np.std(acuracias_classe_0)
    media_acuracia_analise_classe_1 = np.mean(acuracias_classe_1)
    desvio_acuracia_analise_classe_1 = np.std(acuracias_classe_1)
    
    print('\nAnálise Discriminante Gaussiana:')
    print(f'{"-"*65}')
    print(f'Acurácia média: {media_acuracia_analise}')
    print(f'Desvio padrão da acurácia: {desvio_acuracia_analise}')
    print(f'{"-"*65}')
    print(f'Acurácia média para a classe 0: {media_acuracia_analise_classe_0}')
    print(f'Desvio padrão da acurácia para a classe 0: {desvio_acuracia_analise_classe_0}')
    print(f'{"-"*65}')
    print(f'Acurácia média para a classe 1: {media_acuracia_analise_classe_1}')
    print(f'Desvio padrão da acurácia para a classe 1: {desvio_acuracia_analise_classe_1}')
    print(f'{"-"*65}\n')
    
    # Plotar o gráfico
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, 11), acuracias_Análise, marker='o', linestyle='-', color='b')
    plt.xlabel('Fold')
    plt.ylabel('Acurácia')
    plt.title('Acurácia Média em cada Fold')
    plt.grid(True)
    plt.show()
```

# Código para o modelo do Naive Bayes Gaussiano

## a)  Considerando uma validação cruzada em 10 folds, avalie modelos de classificação binária nos dados em questão. Para tanto, use as abordagens abaixo:

```python
# Função para calcular os parâmetros
def calcular_parametros_naive_bayes(X, y):
    phi = np.mean(y)
    mu_0 = np.mean(X[y == 0], axis=0)
    mu_1 = np.mean(X[y == 1], axis=0)
    var_0 = np.var(X[y == 0], axis=0)
    var_1 = np.var(X[y == 1], axis=0)
    return phi, mu_0, mu_1, var_0, var_1

# Função para calcular a densidade de probabilidade gaussiana
def calcular_densidade_probabilidade_gaussiana(x, mu, var, epsilon=1e-8):
    var = np.maximum(var, epsilon)
    coef = 1.0 / np.sqrt(2.0 * np.pi * var)
    exponen = np.exp(-0.5 * ((x - mu) ** 2 / var))
    return coef * exponen

# Função para prever usnado o metodo
def prever_naive_bayes_gaussiano(X, phi, mu_0, mu_1, var_0, var_1):
    p_0 = np.log(1 - phi)
    p_1 = np.log(phi)
    for d in range(X.shape[1]):
        p_0 += np.log(calcular_densidade_probabilidade_gaussiana(X[:, d], mu_0[d], var_0[d]))
        p_1 += np.log(calcular_densidade_probabilidade_gaussiana(X[:, d], mu_1[d], var_1[d]))
    
    res = []
    for i in range(len(p_1)):
        if p_1[i] > p_0[i]:
            res.append(1)
        else:
            res.append(0)
    
    return np.array(res)

# Função principal para realizar o Naive Bayes Gaussiano
def naive_bayes_gaussiano1(file):

    X, y = csv(file)
    X = normalizar_dados(X)

    Indice_KF = kfold_10(X, y)
    acuracias_Naive = []
    acuracias_classe_1 = []
    acuracias_classe_0 = []

    for i in range(10):
        indice_teste = Indice_KF[i]
        indice_treino = np.concatenate([Indice_KF[j] for j in range(10) if j != i])
        
        X_treino, X_teste = X[indice_treino], X[indice_teste]
        y_treino, y_teste = y[indice_treino], y[indice_teste]
        
        # Calcular parâmetros
        phi, mu_0, mu_1, var_0, var_1 = calcular_parametros_naive_bayes(X_treino, y_treino)
        
        # Prever com o Naive Bayes Gaussiano
        y_pred = prever_naive_bayes_gaussiano(X_teste, phi, mu_0, mu_1, var_0, var_1)
        acuracia = np.mean(y_pred == y_teste)
        acuracias_Naive.append(acuracia)
        
        # Acurácia por classe
        acuracia_classe_1 = np.mean(y_pred[y_teste == 1] == 1) if np.any(y_teste == 1) else 0
        acuracia_classe_0 = np.mean(y_pred[y_teste == 0] == 0) if np.any(y_teste == 0) else 0
        acuracias_classe_1.append(acuracia_classe_1)
        acuracias_classe_0.append(acuracia_classe_0)
    
    # Calcular e exibir os resultados
    acuracia_media_Naive = np.mean(acuracias_Naive)
    desvio_padrao_acuracia_Naive = np.std(acuracias_Naive)
    acuracia_media_Naive_classe_1 = np.mean(acuracias_classe_1)
    desvio_padrao_acuracia_Naive_classe_1 = np.std(acuracias_classe_1)
    acuracia_media_Naive_classe_0 = np.mean(acuracias_classe_0)
    desvio_padrao_acuracia_Naive_classe_0 = np.std(acuracias_classe_0)
    
    # Exibir os dados
    print('Naive Bayes Gaussiano:')
    print(f'{"-"*65}')
    print(f'Acurácia média: {acuracia_media_Naive}')
    print(f'Desvio padrão da acurácia: {desvio_padrao_acuracia_Naive}')
    print(f'{"-"*65}')
    print(f'Acurácia média para a classe 0: {acuracia_media_Naive_classe_0}')
    print(f'Desvio padrão da acurácia para a classe 0: {desvio_padrao_acuracia_Naive_classe_0}')
    print(f'{"-"*65}')
    print(f'Acurácia média para a classe 1: {acuracia_media_Naive_classe_1}')
    print(f'Desvio padrão da acurácia para a classe 1: {desvio_padrao_acuracia_Naive_classe_1}')
    print(f'{"-"*65}')
    
    # Plotar Gráfico
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, 11), acuracias_Naive, marker='o', linestyle='-', color='b')
    plt.xlabel('Fold')
    plt.ylabel('Acurácia')
    plt.title('Acurácia Média em cada Fold')
    plt.grid(True)
    plt.show()
```

## Algumas funções que serão usadas nos métodos a seguir:

```python
# ler o CSV
def csv(file):
    data = pd.read_csv(file)
    X = data.iloc[:, :-1].values
    y = data.iloc[:, -1].values
    return X, y

# FUNÇÃO PARA normalizar os dados
def normalizar_dados(X):
    return (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Implementação do K-Fold
def KFold_10(X, y):
    k = 10
    indices = np.arange(X.shape[0])
    np.random.shuffle(indices)
    tam_fold = len(indices) // k
    return [indices[i * tam_fold:(i + 1) * tam_fold] for i in range(k)]
```

# Código para o modelo de Regressão softmax com GD

## a) Considerando uma validação cruzada em 10 folds, avalie modelos de classificação multiclasse nos dados em questão. Para tanto, use as abordagens abaixo:

```python
# Função Softmax
def softmax(z):
    e = np.exp(z - np.max(z, axis=1, keepdims=True))
    return e / np.sum(e, axis=1, keepdims=True)

# Gradiente Descendente para Softmax
def GD_softmax(X, y, alpha, num_itera, num_classes):
    m, n = X.shape
    theta = np.zeros((n, num_classes))
    for _ in range(num_itera):
        h = softmax(X.dot(theta))
        erro = h - np.eye(num_classes)[y]
        gradiente = X.T.dot(erro) / m
        theta -= alpha * gradiente
    return theta

# # Função principal para realizar a Regressão Softmax
def regressao_softmax(arquivo):

    X, y = csv(arquivo)
    X = normalizar_dados(X)

    # Converter as classes
    classes, y_encoded = np.unique(y, return_inverse=True)
    
    # Adicionar coluna de "1"
    X = np.c_[np.ones((X.shape[0], 1)), X]

    indices_folds = KFold_10(X, y_encoded)
    num_classes = 4
    Softmax_acuracias = []
    acuracias_classe = {classe: [] for classe in classes}

    for i in range(len(indices_folds)):
        index_teste = indices_folds[i]
        index_treino = np.concatenate([indices_folds[j] for j in range(len(indices_folds)) if j != i])

        X_treino, X_teste = X[index_treino], X[index_teste]
        y_treino, y_teste = y_encoded[index_treino], y_encoded[index_teste]

        # Treinamento do modelo usando Gradient Descent
        theta = GD_softmax(X_treino, y_treino, alpha=0.01, num_itera=1000, num_classes=num_classes)

        # Avaliação do modelo
        y_pred_prob = softmax(X_teste.dot(theta))
        y_pred = np.argmax(y_pred_prob, axis=1)

        # Acurácia global
        acuracia = np.mean(y_pred == y_teste)
        Softmax_acuracias.append(acuracia)

        # Acurácia por classe
        for classe in classes:
            classe_index = np.where(classes == classe)[0][0]
            indices_classe = np.where(y_teste == classe_index)[0]
            if len(indices_classe) > 0:
                acuracia_classe = np.mean(y_pred[indices_classe] == classe_index)
            else:
                acuracia_classe = 0
            acuracias_classe[classe].append(acuracia_classe)

    # Exibir resultados da Regressão Softmax
    softmax_media_acuracia = np.mean(Softmax_acuracias)
    softmax_desvio_acuracia = np.std(Softmax_acuracias)
    print('Regressão Softmax:')
    print(f'{"-"*65}')
    print(f'Acurácia média: {softmax_media_acuracia}')
    print(f'Desvio padrão da acurácia: {softmax_desvio_acuracia}')
    for classe in classes:
        media_acuracia_classe = np.mean(acuracias_classe[classe])
        desvio_acuracia_classe = np.std(acuracias_classe[classe])
        print(f'{"-"*65}')
        print(f'Acurácia média para a classe {classe}: {media_acuracia_classe}')
        print(f'Desvio padrão da acurácia para a classe {classe}: {desvio_acuracia_classe}')
    
    print(f'{"-"*65}')
    
    # Plotar o gráfico
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, 11), Softmax_acuracias, marker='o', linestyle='-', color='b')
    plt.xlabel('Fold')
    plt.ylabel('Acurácia')
    plt.title('Acurácia Média em cada Fold')
    plt.grid(True)
    plt.show()
```

# Código para o modelo de Análise do discriminante Gaussiano

## a) Considerando uma validação cruzada em 10 folds, avalie modelos de classificação multiclasse nos dados em questão. Para tanto, use as abordagens abaixo:¶

```python
# Função para treinar o modelo usando a Análise do Discriminante Gaussiano (GDA)
def treinar_GDA(X, y, num_classes):
    m, n = X.shape
    med = np.zeros((num_classes, n))
    covar = np.zeros((n, n))
    probab = np.zeros(num_classes)
    
    for classe in range(num_classes):
        X_classe = X[y == classe]
        med[classe] = np.mean(X_classe, axis=0)
        probab[classe] = X_classe.shape[0] / m
    
    covar = np.cov(X.T, bias=True)
    return med, covar, probab

def prever_GDA(X, med, cov, probab):
    num_classes = med.shape[0]
    m = X.shape[0]
    prev = np.zeros(m)
    
    for i in range(m):
        posteriores = np.zeros(num_classes)
        for classe in range(num_classes):
            dif = X[i] - med[classe]
            verossimilhanca = (1 / (np.sqrt((2 * np.pi) ** med.shape[1] * np.linalg.det(cov)))) * \
                              np.exp(-0.5 * dif.dot(np.linalg.inv(cov)).dot(dif))
            posteriores[classe] = probab[classe] * verossimilhanca
        prev[i] = np.argmax(posteriores)
    
    return prev

# Função principal para realizar o GDA
def GDA(arquivo):

    X, y = csv(arquivo)
    X = normalizar_dados(X)

    # Converter as Classes
    classes, y_codificado = np.unique(y, return_inverse=True)

    num_classes = 4
    folder_Indice = KFold_10(X, y_codificado)
    acuracias_gda = []
    acuracias_classe = {classe: [] for classe in classes}

    for i in range(len(folder_Indice)):
        indice_teste = folder_Indice[i]
        indice_treino = np.concatenate([folder_Indice[j] for j in range(len(folder_Indice)) if j != i])

        X_treino, X_teste = X[indice_treino], X[indice_teste]
        y_treino, y_teste = y_codificado[indice_treino], y_codificado[indice_teste]

        # Treinamento do modelo
        med, cov, probab = treinar_GDA(X_treino, y_treino, num_classes)

        # Avaliação
        y_predito = prever_GDA(X_teste, med, cov, probab)

        # Acurácia global
        acuracia = np.mean(y_predito == y_teste)
        acuracias_gda.append(acuracia)

        # Acurácia por classe
        for classe in classes:
            classe_index = np.where(classes == classe)[0][0]
            indices_classe = np.where(y_teste == classe_index)[0]
            if len(indices_classe) > 0:
                acuracia_classe = np.mean(y_predito[indices_classe] == classe_index)
            else:
                acuracia_classe = 0
            acuracias_classe[classe].append(acuracia_classe)

    # Exibir resultados
    media_acuracia_gda = np.mean(acuracias_gda)
    desvio_acuracia_gda = np.std(acuracias_gda)
    print('Análise do Discriminante Gaussiano (GDA):')
    print(f'{"-"*65}')
    print(f'Acurácia média: {media_acuracia_gda}')
    print(f'Desvio padrão da acurácia: {desvio_acuracia_gda}')
    for classe in classes:
        print(f'{"-"*65}')
        media_acuracia_classe = np.mean(acuracias_classe[classe])
        desvio_acuracia_classe = np.std(acuracias_classe[classe])
        print(f'Acurácia média para a classe {classe}: {media_acuracia_classe}')
        print(f'Desvio padrão da acurácia para a classe {classe}: {desvio_acuracia_classe}')
    
    print(f'{"-"*65}')
    
    # Plotar grafico
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, 11), acuracias_gda, marker='o', linestyle='-', color='b')
    plt.xlabel('Fold')
    plt.ylabel('Acurácia')
    plt.title('Acurácia Média em cada Fold (GDA)')
    plt.grid(True)
    plt.show()
```

# Código para o modelo de Naive Bayes Gaussiano

## a) Considerando uma validação cruzada em 10 folds, avalie modelos de classificação multiclasse nos dados em questão. Para tanto, use as abordagens abaixo:¶

```python
# Função para ajustar o modelo Naive Bayes
def ajustar_naive(X, y, num_classes):
    m, n = X.shape
    mu = np.zeros((num_classes, n))
    var = np.zeros((num_classes, n))
    pi = np.zeros(num_classes)
    
    for i in range(num_classes):
        X_aux = X[y == i]
        mu[i] = np.mean(X_aux, axis=0)
        var[i] = np.var(X_aux, axis=0)
        pi[i] = X_aux.shape[0] / m
    
    return mu, var, pi

# Função para prever com o modelo Naive Bayes
def prever_naive(X, mu, var, pi):
    num_classes = mu.shape[0]
    m = X.shape[0]
    y_pred = np.zeros(m)
    
    for i in range(m):
        posteriores = np.zeros(num_classes)
        for classe_aux in range(num_classes):
            prever = np.log(pi[classe_aux])
            k = -0.5 * np.sum(np.log(2 * np.pi * var[classe_aux]))
            k -= 0.5 * np.sum(((X[i] - mu[classe_aux]) ** 2) / var[classe_aux])
            posteriores[classe_aux] = prever + k
        y_pred[i] = np.argmax(posteriores)
    
    return y_pred

# Função principal para realizar o Naive Bayes Gaussiano
def naive_bayes_gaussiano(arquivo):

    X, y = csv(arquivo)
    X = normalizar_dados(X)

    # Converter as classes
    classes, y_encoded = np.unique(y, return_inverse=True)

    num_classes = 4
    indices_folds = KFold_10(X, y_encoded)
    acuracias_naive = []
    acuracias_classe = {classe: [] for classe in classes}

    for i in range(len(indices_folds)):
        index_teste = indices_folds[i]
        index_treino = np.concatenate([indices_folds[j] for j in range(len(indices_folds)) if j != i])

        X_treino, X_teste = X[index_treino], X[index_teste]
        y_treino, y_teste = y_encoded[index_treino], y_encoded[index_teste]

        # Treinamento
        mu, var, pi = ajustar_naive(X_treino, y_treino, num_classes)

        # Avaliação do modelo
        y_pred = prever_naive(X_teste, mu, var, pi)

        # Acurácia global
        acuracia = np.mean(y_pred == y_teste)
        acuracias_naive.append(acuracia)

        # Acurácia por classe
        for classe in classes:
            classe_index = np.where(classes == classe)[0][0]
            indices_classe = np.where(y_teste == classe_index)[0]
            if len(indices_classe) > 0:
                acuracia_classe = np.mean(y_pred[indices_classe] == classe_index)
            else:
                acuracia_classe = 0
            acuracias_classe[classe].append(acuracia_classe)

    # Exibir resultados do Naive Bayes
    media_acuracia_naive = np.mean(acuracias_naive)
    desvio_acuracia_naive = np.std(acuracias_naive)
    print('Naive Bayes:')
    print(f'{"-"*65}')
    print(f'Acurácia média: {media_acuracia_naive}')
    print(f'Desvio padrão da acurácia: {desvio_acuracia_naive}')
    for classe in classes:
        print(f'{"-"*65}')
        media_acuracia_classe = np.mean(acuracias_classe[classe])
        desvio_acuracia_classe = np.std(acuracias_classe[classe])
        print(f'Acurácia média para a classe {classe}: {media_acuracia_classe}')
        print(f'Desvio padrão da acurácia para a classe {classe}: {desvio_acuracia_classe}')

    print(f'{"-"*65}')
    
    # Plotar Gráfico
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, 11), acuracias_naive, marker='o', linestyle='-', color='b')
    plt.xlabel('Fold')
    plt.ylabel('Acurácia')
    plt.title('Acurácia Média em cada Fold (Naive Bayes)')
    plt.grid(True)
    plt.show()
```
