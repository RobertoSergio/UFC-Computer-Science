## Funções que serão usadas nos dois métodos:

```python
# função para ler o arquivo
def ler_csv(arquivo):
    df = pd.read_csv(arquivo)
    X = df.iloc[:, :-1].values
    y = df.iloc[:, -1].values.astype(int)
    return X, y

# Função para normalizar os dados
def normalizar_dados(X):
    return (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Função da implementação do kfold para validação cruzada em 10 folds
def kfold_10(X, y):
    n = X.shape[0]
    indices = np.arange(n)
    np.random.shuffle(indices)
    
    # Tamanho de cada fold
    tam_aux = n // 10
    res = n % 10
    tamanho_fold = [tam_aux + 1] * res + [tam_aux] * (10 - res)

    i = 0
    folds = []
    for tam in tamanho_fold:
        inicio, fim = i, i + tam
        folds.append(indices[inicio:fim])
        i = fim

    return folds

# função para calcular as métricas pedidas no item b
def calcular_metricas(y_true, y_pred):
    acuracia = accuracy_score(y_true, y_pred)
    precisao = precision_score(y_true, y_pred, average='weighted')
    revocacao = recall_score(y_true, y_pred, average='weighted')
    f1 = f1_score(y_true, y_pred, average='weighted')
    return acuracia, precisao, revocacao, f1
```

# Código para o modelo KNN

## a) Considerando uma validação cruzada em 10 folds, avalie modelos de classificação binária nos dados em questão.

```python
# função da distancia euclidiana
def dist_euclidiana(a, b):
    delta = a - b
    return np.sqrt(np.sum((delta) ** 2))

# função da distancia mahalanobis
def dist_mahalanobis(a,b, matrix):
    delta = a - b
    return np.sqrt(delta.dot(matrix).dot(delta.T))

# previsão do modelo KNN
def prev_knn(X_treino, y_treino,X_teste,k_aux, distancia, matrix):
    y_pred = []
    for x_teste in X_teste:
        if distancia == 'mahalanobis':
            distancias = np.array([dist_mahalanobis(x_treino,x_teste, matrix) for x_treino in X_treino])
        else:
            distancias = np.array([dist_euclidiana(x_treino,x_teste) for x_treino in X_treino])
        
        k_indices = distancias.argsort()[:k_aux]
        k_vizinhos = y_treino[k_indices]
        y_pred.append(np.argmax(np.bincount(k_vizinhos.astype(int))))
    return np.array(y_pred)

# função principal para o metodo KNN
def KNN(arquivo):
    X, y = ler_csv(arquivo)
    X = normalizar_dados(X)
    folder = kfold_10(X, y)
    k = [1, 5]
    dist = ['euclidiana', 'mahalanobis']
    
    métricas = {k_aux: {d: [[], [], [], [], [], [], [], []] for d in dist} for k_aux in k}

    for i, test_indices in enumerate(folder):
        train_indices = np.concatenate([folder[j] for j in range(len(folder)) if j != i])
        X_treino, X_teste = X[train_indices], X[test_indices]
        y_treino, y_teste = y[train_indices], y[test_indices]
        
        for k_aux in k:
            for d in dist:
                if d == 'mahalanobis':
                    matrix_cov = np.cov(X_treino, rowvar=False)
                    matrix_cov_inv = np.linalg.inv(matrix_cov)
                else:
                    matrix_cov_inv = None
                
                y_pred = prev_knn(X_treino, y_treino, X_teste, k_aux, d, matrix_cov_inv)
                
                # calcular cada métrica
                acuracia, precisao, revocacao, f1 = calcular_metricas(y_teste, y_pred)
                métricas[k_aux][d][0].append(acuracia)    
                métricas[k_aux][d][1].append(revocacao)   
                métricas[k_aux][d][2].append(precisao)    
                métricas[k_aux][d][3].append(f1) 
                
     # Calcular o desvio padrão de cada métrica
    for i in métricas:
        for d in dist:
            métricas[i][d][4] = np.std(métricas[i][d][0]) 
            métricas[i][d][5] = np.std(métricas[i][d][1])
            métricas[i][d][6] = np.std(métricas[i][d][2])
            métricas[i][d][7] = np.std(métricas[i][d][3])
    
    # Printar
    print('KNN:')
    print(f'{"-"*70}')
    for i in métricas:
        for d in dist:
            KNN_acuracia_media = np.mean(métricas[i][d][0])
            KNN_acuracia_desvio = métricas[i][d][4]
            KNN_revocacao_media = np.mean(métricas[i][d][1])
            KNN_revocacao_desvio = métricas[i][d][5]
            KNN_precisao_media = np.mean(métricas[i][d][2])
            KNN_precisao_desvio = métricas[i][d][6]
            KNN_f1_media = np.mean(métricas[i][d][3])
            KNN_f1_desvio = métricas[i][d][7]

            print(f'\nk={i}, distancia={d}:')
            print(f'{"-"*70}')
            print(f'Acurácia média: {KNN_acuracia_media} ± {KNN_acuracia_desvio}')
            print(f'Revocação média: {KNN_revocacao_media} ± {KNN_revocacao_desvio}')
            print(f'Precisão média: {KNN_precisao_media} ± {KNN_precisao_desvio}')
            print(f'F1-score médio: {KNN_f1_media} ± {KNN_f1_desvio}')
            print(f'{"-"*70}')
```

# Arvore de Decisão

## a) Considerando uma validação cruzada em 10 folds, avalie modelos de classificação binária nos dados em questão.

```python
def arvore_de_decisao(arquivo):
    X, y = ler_csv(arquivo)
    X = normalizar_dados(X)
    
    folder = kfold_10(X, y)
    
    criterios = ['gini', 'entropy']
    métricas = {k: [[], [], [], [], [], [], [], []] for k in criterios}

    for i, test_indices in enumerate(folder):
        train_indices = np.concatenate([folder[j] for j in range(len(folder)) if j != i])
        X_treino, X_teste = X[train_indices], X[test_indices]
        y_treino, y_teste = y[train_indices], y[test_indices]
        
        for k in criterios:
            classificador = DecisionTreeClassifier(criterion=k)
            classificador.fit(X_treino, y_treino)
            y_pred = classificador.predict(X_teste)
            
            # calcular cada métrica
            acuracia, precisao, revocacao, f1 = calcular_metricas(y_teste,y_pred)
            
            métricas[k][0].append(acuracia)    
            métricas[k][1].append(revocacao)   
            métricas[k][2].append(precisao)   
            métricas[k][3].append(f1)   
            
    # Calcular o desvio padrão de cada métrica
    for i in métricas:
        métricas[i][4] = np.std(métricas[i][0])  
        métricas[i][5] = np.std(métricas[i][1])  
        métricas[i][6] = np.std(métricas[i][2])  
        métricas[i][7] = np.std(métricas[i][3])  
    
    # Printar
    print('Árvore de Decisão:')
    print(f'{"-"*70}')
    for i in métricas:
        
        # calcular a média de cada métrica e receber os valores do desvio
        tree_acuracia_media = np.mean(métricas[i][0])
        tree_acuracia_desvio = métricas[i][4]
        tree_revocacao_media = np.mean(métricas[i][1])
        tree_revocacao_desvio = métricas[i][5]
        tree_precisao_media = np.mean(métricas[i][2])
        tree_precisao_desvio = métricas[i][6]
        tree_f1_media = np.mean(métricas[i][3])
        tree_f1_desvio = métricas[i][7]

        print(f'\n{i}:')
        print(f'{"-"*70}')
        print(f'Acurácia média: {tree_acuracia_media} ± {tree_acuracia_desvio}')
        print(f'Revocação média: {tree_revocacao_media} ± {tree_revocacao_desvio}')
        print(f'Precisão média: {tree_precisao_media} ± {tree_precisao_desvio}')
        print(f'F1-score médio: {tree_f1_media} ± {tree_f1_desvio}')
        print(f'{"-"*70}')
```

##  b) Para cada modelo criado, reporte valor médio e desvio padrão das métricas de acurácia, revocação, precisão e F1-score:
