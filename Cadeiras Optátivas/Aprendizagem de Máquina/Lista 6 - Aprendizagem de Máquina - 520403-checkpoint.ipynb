## Implementação do K-médias ( Infelizmente, a parte da repetição só foi feita 5 vezes para não demorar muito)

```python
# Função para calcular a distância Euclidiana
def dist_euclidiana(a, b):
    delta = a - b
    return np.sqrt(np.sum(delta ** 2))

# Função para calcular a distância de Mahalanobis
def dist_mahalanobis(a, b, matrix):
    delta = a - b
    return np.sqrt(delta.dot(matrix).dot(delta.T))

# inicialização dos centroids
def inicializar_centroids(X, k):
    np.random.seed(42)
    return X[np.random.choice(X.shape[0], k, replace=False)]

#atualização dos centroids
def atualizar_centroids(X, cluster_labels, k):
    lista_novos_centroids = []
    for i in range(k):
        pontos_cluster = X[cluster_labels == i]
        if len(pontos_cluster) > 0:
            lista_novos_centroids.append(np.mean(pontos_cluster, axis=0))
        else:
            lista_novos_centroids.append(np.random.randn(1, X.shape[1])) 
    return np.array(lista_novos_centroids)

# Atribuir cada ponto ao cluster mais próximo
def clusters(X, centroids, função_distancia, matrix=None):
    lista_cluster = []
    for aux in X:
        distancia = [função_distancia(aux, centroid, matrix) if matrix is not None else função_distancia(aux, centroid) for centroid in centroids]
        lista_cluster.append(np.argmin(distancia))
    return np.array(lista_cluster)

# Erro de reconstrução
def erro_reconstrucao(X, lista_cluster, centroids):
    inercia = 0
    for i in range(len(centroids)):
        pontos_cluster = X[lista_cluster == i]
        inercia += np.sum(np.linalg.norm(pontos_cluster - centroids[i], axis=1)**2)
    return inercia

# Implementação do K-médias com múltiplas execuções
def k_media(X, k, função_distancia, num_repeticoes, max_iters=100, matrix=None):
    melhor_inercia = np.inf
    melhor_cluster = None
    melhor_centroids = None
    
    for _ in range(num_repeticoes):
        centroids = inicializar_centroids(X, k)
        for _ in range(max_iters):
            lista_cluster = clusters(X, centroids, função_distancia, matrix)
            lista_novos_centroids = atualizar_centroids(X, lista_cluster, k)
            if np.all(centroids == lista_novos_centroids):
                break
            centroids = lista_novos_centroids

        inercia_atual = erro_reconstrucao(X, lista_cluster, centroids)
        
        # Verificar se esta execução foi melhor
        if inercia_atual < melhor_inercia:
            melhor_inercia = inercia_atual
            melhor_cluster = lista_cluster
            melhor_centroids = centroids
    
    return melhor_cluster, melhor_centroids, melhor_inercia

# Calculo das distâncias intra-cluster
def dist_intra_cluster(X, lista_cluster, centroids, k):
    vetor_dist_intra_cluster = np.zeros(k)
    
    for i in range(k):
        pontos_cluster = X[lista_cluster == i]
        if len(pontos_cluster) > 0:
            vetor_dist_intra_cluster[i] = np.mean([np.linalg.norm(aux - centroids[i]) for aux in pontos_cluster])
        else:
            vetor_dist_intra_cluster[i] = 0
    
    return vetor_dist_intra_cluster

# Calculo das distâncias entre os centróides
def inter_cluster_distances(centroids, k):
    dist_intra_aux = np.zeros((k, k))
    for i in range(k):
        for j in range(k):
            if i != j:
                dist_intra_aux[i, j] = dist_euclidiana(centroids[i], centroids[j])
    return dist_intra_aux

# Calculo do índice de Davies-Bouldin
def Davies_Bouldin(X, lista_cluster, centroids):
    k = len(centroids)
    vetor_dist_intra_cluster = dist_intra_cluster(X, lista_cluster, centroids, k)
    dist_intra_aux = inter_cluster_distances(centroids, k)
    
    index_db = np.zeros(k)
    
    for i in range(k):
        max_razao_disp = -np.inf
        for j in range(k):
            if i != j:
                razao_disp = (vetor_dist_intra_cluster[i] + vetor_dist_intra_cluster[j]) / dist_intra_aux[i, j]
                if razao_disp > max_razao_disp:
                    max_razao_disp = razao_disp
        index_db[i] = max_razao_disp
    
    return np.mean(index_db)

# Função main para encontrar o melhor K com base no índice DB
# esses valores min e max de k são da propria lista
def Melhor_k(X, função_distancia, num_repeticoes, min_k=4, max_k=20, matrix=None):
    melhor_k = None
    melhor_index_db = np.inf
    valores_db = []
    
    for k in range(min_k, max_k + 1):
        lista_cluster, centroids, _ = k_media(X, k, função_distancia, num_repeticoes=num_repeticoes, matrix=matrix)
        index_db = Davies_Bouldin(X, lista_cluster, centroids)
        valores_db.append(index_db)
        
        if index_db < melhor_index_db:
            melhor_index_db = index_db
            melhor_k = k
    
    return melhor_k, valores_db

# main
def funcao_principal_k_media(data):
    X = data.values

    # Matriz inversa -> distância de Mahalanobis
    VI = np.linalg.inv(np.cov(X.T))

    # a)
    best_k_euclidean, db_scores_euclidean = Melhor_k(X, dist_euclidiana, num_repeticoes=5)
    print(f"Melhor K com distância Euclidiana: {best_k_euclidean}")

    # b) 
    best_k_mahalanobis, db_scores_mahalanobis = Melhor_k(X, dist_mahalanobis, matrix=VI, num_repeticoes=5)
    print(f"Melhor K com distância de Mahalanobis: {best_k_mahalanobis}")

# Chamando a main
funcao_principal_k_media(data)
```

## Implementação do PCA para 2 dimensões

```python
# Separar atributos e classes
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

# Função para normalizar os dados
def normalizar_dados(X):
    return (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Normalizando os dados
X_normalizado = normalizar_dados(X)

def pca(X, n_components):
    cov_matrix = np.cov(X.T)
    autovalores, autovetores = np.linalg.eig(cov_matrix)
    
    indices = np.argsort(autovalores)[::-1]
    autovalores = autovalores[indices]
    autovetores = autovetores[:, indices]
    
    
    componentes = autovetores[:, :n_components]
    X_pca = X.dot(componentes)
    return X_pca

# Projeção em 2 dimensões
X_pca = pca(X_normalizado, n_components=2)

# Plotar a projeção
plt.figure(figsize=(8, 6))
for especie in np.unique(y):
    plt.scatter(X_pca[y == especie, 0], X_pca[y == especie, 1], label=f'Espécie {especie}')
plt.title('PCA - Projeção em 2 Dimensões')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend()
plt.grid()
plt.show()
```
